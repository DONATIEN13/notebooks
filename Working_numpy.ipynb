#Exercice: Zoomer une photo chargÃ©e automatiquement dans scipy de 1/4 vers le milieu de cette photo 

```python
from scipy import misc
import matplotlib.pyplot as plt
face = misc.face(gray = True)
plt.imshow(face, cmap = plt.cm.gray)
plt.show()
```

```python
face.shape
```

```python
hauteur = face.shape[0]
largeur = face.shape[1]

zoom_face = face[hauteur//4 : -hauteur//4, largeur//4 : -largeur//4]
zoom_face[zoom_face > 155 ] = 250
zoom_face[zoom_face < 155 ] = 0
plt.imshow(zoom_face, cmap = plt.cm.gray)
plt.show()
```

# OPTIMISATION AVEC curve_fit

```python
from scipy import optimize
x = np.linspace(0, 2, 100)
y = 1/3*x**3 - 3/5*x**2 + 2 + np.random.randn(x.shape[0])/20
plt.scatter(x, y)
```

```python
def f(x, a, b ,c, d):
    return a*x**3 + b*x**2 + c*x + d
```

```python
params, paramcov = optimize.curve_fit(f, x, y)
plt.scatter(x, y)
plt.plot(x, f(x, params[0], params[1], params[2], params[3]), c = 'g', lw = 3)
```

#### Minimisation

```python
def f(x):
    return x**2 + 15*np.sin(x)
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, f(x))
```

```python
result = optimize.minimize(f, x0 = -5).x
```

```python
plt.plot(x, f(x), lw = 3, zorder = -1)
plt.scatter(result, f(result), s = 100, c = 'r', zorder = 1)
```

# Transformation de Fourier

```python
x = np.linspace(0, 30, 1000)
y = 3*np.sin(x) + 2*np.sin(5*x) + np.sin(10*x) + np.random.random(x.shape[0])*10
plt.plot(x,y)
```

```python
fourier = fftpack.fft(y)
power = np.abs(fourier)
frequences = fftpack.fftfreq(y.size)
plt.plot(np.abs(frequences), power)
```

```python
fourier[power<400] = 0
```

```python
plt.plot(np.abs(frequences), np.abs(fourier))
```

```python
filtered_signal = fftpack.ifft(fourier)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(x, y, lw = 0.5)
plt.plot(x, filtered_signal, lw = 3)
```

#### Exercice: Modifier la colonne 'age' pour creer 04 cathegories de personnes(moins de 20 ans, entre 20 et 30 ans, entre 30 et 40 ans et plus de 40 ans)

```python
def cathe(age):
    if age <= 20:
        return '<20ans'
    elif (age > 20) & (age <= 30):
        return '20-30ans'
    elif (age > 30) & (age <= 40):
        return '30-40ans'
    else:
        return '+40ans'
```

```python
data['age'].map(cathe)
```
