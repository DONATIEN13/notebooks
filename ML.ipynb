```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
```

### 1) REGRESSION LINEAIRE SIMPLE

```python
np.random.seed(0)

m = 50 
X = np.random.randn(m).reshape(-1, 1)
y = 1.5 * X + 3 + np.random.randn(m).reshape(-1, 1)

plt.scatter(X, y)
```

```python
model = LinearRegression()
model.fit(X, y)

new_data = np.linspace(-3, 3).reshape(-1, 1)
predictions = model.predict(new_data)

plt.scatter(X, y)
plt.plot(new_data, predictions, c = 'green')
```

### 2) CLASSIFICATION

```python
X, y = make_blobs(n_samples = 50,
                 centers = 2,
                 n_features = 2,
                 random_state = 0,
                 cluster_std = 1)

X_new = np.random.uniform(low = X.min(), 
                          high= X.max(), 
                          size = (10,2))


plt.scatter(X[:, 0], X[:, 1], c = y, cmap = "bwr")
plt.scatter(X_new[:, 0], X_new[:, 0], c = 'white', edgecolor = 'k')
```

```python
model = KNeighborsClassifier()
model.fit(X, y)

predictions = model.predict(X_new)
plt.scatter(X[:, 0], X[:, 1], alpha = 0.4, c = y)
plt.scatter(X_new[:, 0], X_new[:, 1], c = predictions, edgecolor = 'k')
```

### 3) CLUSTERING

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
```

```python
k = np.random.choice([2, 3, 4, 5, 6])

X, _ = make_blobs(n_samples = 200, centers = k, n_features = 2, cluster_std = 0.5)
```

```python
k
```

```python
plt.scatter(X[:, 0], X[:, 1], c = _, alpha = 0.4)
```

##### * La méthode du coude pour determiner le nombre de cluster qu'il faut créer

```python
k_range = range(1, 10)
k_inertias = []

for k in k_range:
    model = KMeans(n_clusters = k)
    model.fit(X)
    k_inertias.append(model.inertia_)

plt.plot(k_range, k_inertias, lw = 2)
```

```python
model = KMeans(n_clusters = 4)
model.fit(X)

predictions = model.predict(X)

plt.scatter(X[:, 0], X[:, 1], c = predictions, alpha = 0.3)
```

#### DETECTION D'ANOMALIES

```python
import numpy as np
import matplotlib.pyplot as plt
#from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_blobs
from sklearn.neighbors import LocalOutlierFactor
```

```python
np.random.seed(0)

X = np.random.randn(30,2)*0.2
X = np.concatenate((X, np.array([[2, 0.5]])), axis = 0)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.xlim(-2, 3)
plt.ylim(-1, 1)
```

```python
model = LocalOutlierFactor(n_neighbors = 20, contamination = 0.5)
predictions = model.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c = predictions, cmap = "bwr")
plt.xlim(-2, 3)
plt.ylim(-1, 1)
```
